```python
from PIL import Image


def decode(image,length):
    pixels = image.load()
    bin_data = []
    current_string = ""
    iteration = 0
    
    for i in range(image.size[0]):
        for j in range(image.size[1]):
            for p in range(len(pixels[i,j])):
                if pixels[i,j][p]%2==0:
                    current_string += "0"
                else:
                    current_string += "1"
                if len(current_string) == 8:
                    bin_data.append(current_string)
                    current_string = ""
                iteration +=1
                if iteration == length:
                    print(f"length:{length} and iteration: {iteration}")
                    return bin_data
    

def encode(text,image):
    #image.show()
    #for im in image.getdata():
        #print(im)
        
    pixels = image.load()
    
    text = ''.join(format(x,'b') for x in bytes(text,"ascii"))
    print(text)
    iteration = 0

    for i in range(image.size[0]):
        for j in range(image.size[1]):
            
            tmp = list(pixels[i,j])
            
            for k in range(len(tmp)):
                if iteration == len(text)-1:
                    return image , len(text)
                
                elif text[iteration]=="0":
                    if tmp[k]%2!=0:
                        if tmp[k]==255:
                            tmp[k]-=1
                        else:
                            tmp[k]+=1
                            
                            
                elif text[iteration]=="1":
                    if tmp[k]%2==0:
                        tmp[k]+=1
                        
                iteration +=1
                
            image.putpixel((i,j),tuple(tmp))
            print(f" {i} {j} pixel : {pixels[i,j]} , tmp: {tmp}")
    
            
                        

```
